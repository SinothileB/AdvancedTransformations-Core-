```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json
```

```python
df = pd.read_csv('superhero_powers.csv')
df.info()
df.head()
```

```python
df2 = pd.read_csv('superhero_info.csv')
df2.info()
df2.head()
```

```python
## Splits the columns into 2 
df2['Hero|Publisher'].str.split('|',expand=True)
```

```python
## save the 2 new columns into the dataframe
df2[['Hero','Publisher']] = df2['Hero|Publisher'].str.split('|',expand=True)
df2.head(2)
```

```python
## drop the original column 
df2 = df2.drop(columns=['Hero|Publisher'])
df2.head(2)
```

```python
### Splits the heigh and weight to 2
## examining a single value from the measurements col
mes = df2.loc[0,"Measurements"]
print(type(mes))
mes
```

```python
import json
json.loads(mes)
```

```python
mes = mes.replace("'",'"')
mes
```

```python
## now we can use json.loads
fixed_mes = json.loads(mes)
print(type(fixed_mes))
fixed_mes
```

```python
## use .str.replace to replace all single quotes
df2['Measurements'] = df2['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
df2['Measurements'] = df2['Measurements'].apply(json.loads)
df2['Measurements'].head()
```

```python
## check a single value after transformation
test_mes= df2.loc[0, 'Measurements']
print(type(test_mes))
test_mes
```

```python
# Unpack a column of dictionaries into separate columns
Hght_wght =df2['Measurements'].apply(pd.Series)
Hght_wght
```

```python
# concat hgt_wght with original dataframe
df2 = pd.concat((df2, Hght_wght), axis = 1)
df2.head(2)
```

```python
#Drop measurements
df2 = df2.drop(columns=['Measurements'])
df2.head(2)
```

```python
#Make height ,weight columns numeric
## save the 2 new columns into the dataframe
df2[['height','cm']] = df2['Height'].str.split(' ',expand=True)
df2.head(2)
```

```python
#Drop 
df2 = df2.drop(columns=['cm'])
df2.head(2)
```

```python
## save the 2 new columns into the dataframe
df2[['weight','kg']] = df2['Weight'].str.split(' ',expand=True)
df2.head(2)
```

```python
#Drop 
df2 = df2.drop(columns=['kg'])
df2.head(2)
```

```python
#Drop measurements
df2 = df2.drop(columns=['Height','Weight'])
df2.head(2)
```
