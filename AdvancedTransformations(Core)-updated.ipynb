```python
#one-hot-encoded columns for every power that appears in the dataset
df.head(2)
```

```python
## showing the lists are really strings
df.loc[2,'Powers']
```

```python
df['Powers_split'] = df['Powers'].str.split(',')
df
```

```python
exploded_p = df.explode('Powers_split')
exploded_p
```

```python
#Now we will take the .unique values from this exploded column as
#our list of columns to create from our original non-exploded dataframe.
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded_p['Powers_split'].dropna().unique()
cols_to_make
```

```python
#We now have a single list that includes ALL unique values,
#found anywhere in the lists within the original transaction column.
```

```python
#Now we want to create a new column for each of our unique values that are stored in cols_to_make.   
#And while we are at it,we want to indicate true and false depending on if the value was in the original list
```

```python
for col in cols_to_make:
    df[col] = df['Powers'].str.contains(col)
df.head()
```

```python
#And now we have our new transaction columns!
#Let's drop the original transactions column and our Powers_split column now that we have confirmed our new columns
df = df.drop(columns=['Powers','Powers_split'])
```
